```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Load the datasets from D:\Jp
trader_data = pd.read_csv(r'D:\Jp\historical_data.csv')
sentiment_data = pd.read_csv(r'D:\Jp\fear_greed_index.csv')

# Print first few rows to verify the data
print(trader_data.head())
print(sentiment_data.head())
```

```python
# Structure and data types
print(trader_data.info())
print(sentiment_data.info())

# Check for missing values
print(trader_data.isnull().sum())
print(sentiment_data.isnull().sum())
```

```python
# Step 4: Clean and Prepare for Merge
trader_data.columns = trader_data.columns.str.strip()
sentiment_data.columns = sentiment_data.columns.str.strip()

# Convert trader timestamps
trader_data['Timestamp IST'] = pd.to_datetime(trader_data['Timestamp IST'], format='%d-%m-%Y %H:%M')
trader_data['date'] = trader_data['Timestamp IST'].dt.date

# Convert sentiment dates
sentiment_data['date'] = pd.to_datetime(sentiment_data['date']).dt.date
```

```python
# Step 5: Merge the Datasets
merged_df = pd.merge(trader_data, sentiment_data, on='date', how='inner')

# Preview merged data
merged_df.head()
```

```python
# Boxplot: Profit vs Sentiment
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.boxplot(x='classification', y='Closed PnL', data=merged_df)
plt.title('Trader Profit Distribution by Market Sentiment')
plt.ylabel('Closed PnL')
plt.xlabel('Market Sentiment')
plt.show()

# Countplot: Number of trades on Fear vs Greed days
plt.figure(figsize=(6, 4))
sns.countplot(x='classification', data=merged_df)
plt.title('Number of Trades on Fear vs Greed Days')
plt.xlabel('Market Sentiment')
plt.ylabel('Number of Trades')
plt.show()
```

```python
# Step 6: Analysis

# Average PnL by sentiment
avg_pnl = merged_df.groupby('classification')['Closed PnL'].mean()
print("Average Profit by Sentiment:\n", avg_pnl)

# Trade count by sentiment
trade_count = merged_df['classification'].value_counts()
print("\nNumber of Trades by Sentiment:\n", trade_count)

# Average PnL by Coin and Sentiment
coin_pnl = merged_df.groupby(['Coin', 'classification'])['Closed PnL'].mean().unstack()
print("\nCoin-wise PnL:\n", coin_pnl)
```

```python
# Step 7: Visualization - Boxplot
sns.boxplot(data=merged_df, x='classification', y='Closed PnL')
plt.title("Profit Distribution by Sentiment")
plt.show()
```

```python
# Step 8: Visualization - Trade Count
sns.countplot(data=merged_df, x='classification')
plt.title("Number of Trades by Sentiment")
plt.show()
```

```python
# Step 9: Visualization - Heatmap by Coin and Sentiment
plt.figure(figsize=(10, 6))
sns.heatmap(coin_pnl, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Average PnL by Coin and Sentiment")
plt.show()
```

## Trader Behavior Insights  Market Sentiment Analysis

### Objective:
Analyze how crypto traders behave under different market sentiments  Fear and Greed  using historical trading data combined with the Fear & Greed Index.

### Key Findings:

#### 1. Profit:
- Traders had higher average profits during **Greed** days.

#### 2. Trade Volume:
- More trades occurred on **Fear** days.

#### 3. Coin Insights:
- Coins like BTC/ETH performed better during **Greed** periods.

### Visuals:
- Boxplot of profit distribution
- Trade count by sentiment
- Heatmap of coin performance vs sentiment

### Conclusion:
Market sentiment impacts trader behavior significantly. Greed days yield more profits, while Fear triggers more frequent trading with lower returns. This insight can support sentiment-aware trading strategies.
